```python
df1['ArrFlightDate'] = df1['FlightDate']
df1['ArrFlightDate'] = pd.to_datetime(df1['ArrFlightDate'])
df1['FlightDate'] = pd.to_datetime(df1['FlightDate'])
```

```python
df1['ArrAbsTime'] = df1['CRSArrTime']
for index, row in df1.iterrows():
    remainder = row["ArrAbsTime"]%100
    absoluteTime = row["ArrAbsTime"] - remainder
    df1.at[index,"ArrAbsTime"] = absoluteTime
    if(row["CRSDepTime"] > row["CRSArrTime"]):
        df1.at[index,"ArrFlightDate"] = row["FlightDate"] + pd.DateOffset(days=1)
```

```python
df1['ArrAbsTime'] = df1['ArrAbsTime'].astype(np.int64)
```

```python
for index,row in df1.iterrows():
    if(row["CRSArrTime"]==2400):
        df1.at[index,"CRSArrTime"]=0
        df1.at[index,"ArrAbsTime"]=0
```

```python
df1["DepAbsTime"] = df1["CRSDepTime"]
for index, row in df1.iterrows():
    remainder = row["DepAbsTime"]%100
    absolutetime = row["DepAbsTime"] - remainder
    df1.at[index,"DepAbsTime"] = absolutetime
```

```python
df1['DepAbsTime'] = df1['DepAbsTime'].astype(np.int64)
```

```python
df2.date = pd.to_datetime(df2.date)
```

```python
merge1 = df1.merge(df2,left_on=["FlightDate","DepAbsTime","Origin"],right_on=["date","time","airport"])
```

```python
merge2 = merge1.merge(df2,left_on=["ArrFlightDate","ArrAbsTime","Dest"],right_on=["date","time","airport"])
```
