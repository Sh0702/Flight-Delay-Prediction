```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
import pandas as pd
from sklearn.model_selection import train_test_split  
from imblearn.under_sampling import RandomUnderSampler,NearMiss
import collections
```

```python
f = open("/Users/shreyassrinivasan/Desktop/SF_Project_Alias/merge_final.csv")
df = pd.read_csv(f)
```

```python
df = df.dropna()
```

```python
x = df.drop(['Unnamed: 0','Unnamed: 0_x','DepDel15','ArrTime','ArrDelayMinutes','ArrDel15',
            'ArrFlightDate','ArrAbsTime','DepAbsTime','Unnamed: 0_y','date_x','time_x','airport_x','Unnamed: 0.1',
            'date_y','time_y','airport_y','FlightDate','OriginAirportID','DestAirportID','DepTime'],axis=1)
```

```python
flightOrigin = {'DFW':1,'EWR':2,'IAH':3,'JFK':4,'LAS':5,'LAX':6,'MCO':7,'MIA':8,'ORD':9,'PHX':10,'SEA':11,'SFO':12,'ATL':13,'CLT':14,'DEN':15} 
```

```python
arr1 = [df['Origin'].apply(lambda x: flightOrigin.get(x))]
```

```python
arr2 = [df['Dest'].apply(lambda x: flightOrigin.get(x))]
```

```python
x['newOrigin'] = arr1[0]
```

```python
x['newDest'] = arr2[0]
```

```python
x = x.drop(['Origin','Dest'],axis=1)
```

```python
x.transpose()
```

```python
y = df['ArrDel15']
```

```python
x.shape
```

```python
y.shape
```

```python
np.any(np.isnan(x))
```

```python
np.all(np.isfinite(x))
```

```python
# Splitting data into training and testing data

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 40)
```

```python
rus = RandomUnderSampler(random_state = 42)
x_resampled1,y_resampled1 = rus.fit_resample(x_train,y_train)
```

```python
nm = NearMiss()
x_resampled2,y_resampled2 = nm.fit_resample(x_train,y_train)
```

```python
print("Undersampling using RandomUnderSampler: "+ str(collections.Counter(y_resampled1)))
print("Undersampling using NearMiss: "+ str(collections.Counter(y_resampled2)))
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression

logReg = LogisticRegression()
```

```python
logReg.fit(x_resampled1,y_resampled1)
```

```python
# Accuracy

score = logReg.score(x_test,y_test)
print(score)
```

```python
pred = logReg.predict(x_test)
```

```python
# Confusion matrix

cm = confusion_matrix(y_test,pred)
```

```python
cm
```

```python
# Classification Report

target_names=["Class 0","Class 1"]
print(classification_report(y_test,pred,target_names=target_names))
```
