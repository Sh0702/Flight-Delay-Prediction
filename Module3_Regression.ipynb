```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
f = open("/Users/shreyassrinivasan/Desktop/SF_Project_Alias/merge_final.csv")
df = pd.read_csv(f)
```

```python
df = df.dropna()
```

```python
x = df.drop(['Unnamed: 0','Unnamed: 0_x','DepDel15','ArrTime','ArrDelayMinutes','ArrDel15',
            'ArrFlightDate','ArrAbsTime','DepAbsTime','Unnamed: 0_y','date_x','time_x','airport_x','Unnamed: 0.1',
            'date_y','time_y','airport_y','FlightDate','OriginAirportID','DestAirportID','DepTime'],axis=1)
```

```python
x.transpose()
```

```python
flightOrigin = {'DFW':1,'EWR':2,'IAH':3,'JFK':4,'LAS':5,'LAX':6,'MCO':7,'MIA':8,'ORD':9,'PHX':10,'SEA':11,'SFO':12,'ATL':13,'CLT':14,'DEN':15} 
```

```python
arr1 = [df['Origin'].apply(lambda x: flightOrigin.get(x))]
```

```python
arr2 = [df['Dest'].apply(lambda x: flightOrigin.get(x))]
```

```python
x['newOrigin'] = arr1[0]
```

```python
x['newDest'] = arr2[0]
```

```python
x = x.drop(['Origin','Dest'],axis=1)
```

```python
y = df['ArrDelayMinutes']
```

```python
x = x.loc[df['ArrDel15'] == 1]
```

```python
arr = []
for i in x.index:
    arr.append(y[i])
y = pd.Series(arr)
```

```python
x.transpose()
```

```python
x.shape
```

```python
y.shape
```

```python
# General Solution
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20,random_state = 42)
```

```python
# Linear Regression

from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
```

```python
linReg.fit(x_train,y_train)
```

```python
pred_linReg = linReg.predict(x_test)
```

```python
print("R2: " , r2_score(y_test,pred_linReg))
print("Root Mean Square Error: ", np.sqrt(mean_squared_error(y_test,pred_linReg)))
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_linReg))
```

```python
# Extra Tree Regressor

from sklearn.ensemble import ExtraTreesRegressor
extTree = ExtraTreesRegressor(n_estimators=100, random_state = 42)
```

```python
extTree.fit(x_train,y_train)
```

```python
pred_extTree = extTree.predict(x_test)
```

```python
print("R2: " , r2_score(y_test,pred_extTree))
print("Root Mean Square Error: ", np.sqrt(mean_squared_error(y_test,pred_extTree)))
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_extTree))
```

```python
# XG Boost Regressor

from sklearn.ensemble import GradientBoostingRegressor
```

```python
xgb = GradientBoostingRegressor(random_state = 42)
```

```python
xgb.fit(x_train,y_train)
```

```python
pred_xgb = xgb.predict(x_test)
```

```python
print("R2: " , r2_score(y_test,pred_xgb))
print("Root Mean Square Error: ", np.sqrt(mean_squared_error(y_test,pred_xgb)))
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_xgb))
```

```python
# Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor
```
