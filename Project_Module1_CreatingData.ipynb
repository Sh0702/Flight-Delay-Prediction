```python
path_weather = "/Users/shreyassrinivasan/Desktop/SF_Project/Weather_Data/"
w_2016 = []
w_2017 = []
for i in os.listdir(path_weather):
    if(i != ".DS_Store"):
        #print(i)
        for j in os.listdir(path_weather + i):
            #print(j)
            if('2016' in j):
                temp = js.load(open(path_weather + i + "/" + j))
                for k in temp["data"]["weather"]:
                    tempdata = pd.DataFrame(k['hourly'],columns = ['windspeedKmph','winddirDegree','weatherCode','precipMM','visibility','pressure','cloudcover','DewPointF','WindGustKmph','tempF','WindChillF','humidity','date','time','airport'])
                    tempdata['date'] = k['date']
                    tempdata['airport'] = i
                    w_2016.append(tempdata)
            if('2017' in j):
                temp = js.load(open(path_weather + i + "/" + j))
                for k in temp["data"]["weather"]:
                    tempdata = pd.DataFrame(k['hourly'],columns = ['windspeedKmph','winddirDegree','weatherCode','precipMM','visibility','pressure','cloudcover','DewPointF','WindGustKmph','tempF','WindChillF','humidity','date','time','airport'])
                    tempdata['date'] = k['date']
                    tempdata['airport'] = i
                    w_2017.append(tempdata)
```

```python
wdf_2016 = pd.concat(w_2016,ignore_index = True)
wdf_2017 = pd.concat(w_2017,ignore_index = True)
```
