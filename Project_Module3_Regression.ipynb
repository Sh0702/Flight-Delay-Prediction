```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
```

```python
f = open("/Users/shreyassrinivasan/Desktop/SF_Project/newData.csv")
df = pd.read_csv(f)
```

```python
df.transpose()
```

```python
X = df.loc[df['ArrDel15'] == 1]
```

```python
X.transpose()
```

```python
y = X['ArrDelayMinutes']
```

```python
X = X.drop(['Unnamed: 0','ArrDelayMinutes','ArrDel15'],axis=1)
```

```python
X.transpose()
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state = 42)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
# Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
linReg = LinearRegression()
linReg.fit(x_train,y_train)
```

```python
pred_linReg = linReg.predict(x_test)
```

```python
len(pred_linReg)
```

```python
print("R2: " , r2_score(y_test,pred_linReg))
print("Root Mean Square Error: ", np.sqrt(mean_squared_error(y_test,pred_linReg)))
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_linReg))
```

```python
# Extra Tree Regressor

from sklearn.ensemble import ExtraTreesRegressor
```

```python
extTree = ExtraTreesRegressor(n_estimators=100, random_state = 42)
extTree.fit(x_train,y_train)
```

```python
pred_extTree = extTree.predict(x_test)
```

```python
print("R2: " , r2_score(y_test,pred_extTree))
print("Root Mean Square Error: ", np.sqrt(mean_squared_error(y_test,pred_extTree)))
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_extTree))
```

```python
# XG Boost Regressor

from sklearn.ensemble import GradientBoostingRegressor
```

```python
xgb = GradientBoostingRegressor(random_state = 42)
xgb.fit(x_train,y_train)
```

```python
pred_xgb = xgb.predict(x_test)
```

```python
print("R2: " , r2_score(y_test,pred_xgb))
print("Root Mean Square Error: ", np.sqrt(mean_squared_error(y_test,pred_xgb)))
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_xgb))
```

```python
# Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor
```

```python
ranFor = RandomForestRegressor(random_state = 12)
ranFor.fit(x_train,y_train)
```

```python
pred_ranFor = ranFor.predict(x_test)
```

```python
print("R2: " , r2_score(y_test,pred_ranFor))
print("Mean Square Error: ", mean_squared_error(y_test,pred_ranFor))
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_ranFor))
```

```python
# RidgeCV

from sklearn.linear_model import RidgeCV
```

```python
ridgeCV = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1])
ridgeCV.fit(x_train,y_train)
```

```python
pred_ridgeCV = ridgeCV.predict(x_test)
```

```python
print("R2: " , r2_score(y_test,pred_ridgeCV))
print("Root Mean Square Error: ", np.sqrt(mean_squared_error(y_test,pred_ridgeCV)))
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_ridgeCV))
```
