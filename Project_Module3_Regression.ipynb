```python
# My Solution
```

```python
"""
predDelay = predicted flight time - expected flight time
predicted flight time = CRSArrTime - DepTime
expected flight time = ArrTime - DepTime

For negative or zero delays(flights which came on time or earlier), predDelay is set to 0. 
"""
```

```python
predFlightTime = [(x1 - x2) for (x1,x2) in zip(df['CRSArrTime'],df['DepTime'])]
```

```python
expFlightTime = [(x1 - x2) for (x1,x2) in zip(df['ArrTime'],df['DepTime'])]
```

```python
predDelay = [(x1 - x2) for (x1,x2) in zip(predFlightTime,expFlightTime)]
```

```python
Y = pd.Series(predDelay)
```

```python
Y.shape
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20,random_state = 42)
```

```python
# Linear Regression

linReg = LinearRegression()
```
